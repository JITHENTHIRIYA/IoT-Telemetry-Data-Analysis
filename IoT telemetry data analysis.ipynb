```python
df1 = df #to not change the original dataframe
def resultValues(x):
    if(x['temp']>=0 and x['temp']<=10):
        return 1
    elif(x['temp']>10 and x['temp']<=20):
        return 2
    else:
        return 3
df1['temp'] = df1.apply(resultValues,axis=1)
df1.head()
```

```python
test = df1.groupby(['temp'])
test.size()
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(num=None, figsize=(7, 7), dpi=80, facecolor='w', edgecolor='k')

corr = df1.corr()

sn.heatmap(corr,annot=True)

plt.show()
```

```python
result_ozone = []
ozone1= []
for i in range(1,len(df1["co"])):
    if df1["co"][i] != 0:
        if df1["temp"][i] != 0:
            ozone1.append(df1["co"][i])
            result_ozone.append(df1["temp"][i])
corr_1 = pd.DataFrame({'co':ozone1,"temp":result_ozone})
corr_1["co"].corr(corr_1["temp"])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
x = df1[['co','light','humidity']]
y = df1['temp']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.40)
```

```python
model2 = RandomForestClassifier()
model2.fit(X_train, y_train)
y_pred= model2.predict(X_test)
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred)*100)
```
